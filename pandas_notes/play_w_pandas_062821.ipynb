```python
my_exp(2, 10)
```

```python
def print_me(x):
    print(x)
    return x
```

```python
df.apply(print_me)
```

```python
import numpy as np
```

```python
def avg_3_apply(col):
    return np.mean(col)
```

```python
avg_3_apply(df)
```

```python
df.apply(avg_3_apply)
```

```python
@np.vectorize

def avg_2_mod(x, y):
    if (x == 20):
        return np.NaN
    else:
        return (x + y)/2
```

```python
df
```

```python
avg_2_mod_vec = np.vectorize(avg_2_mod)
```

```python
avg_2_mod_vec(df['a'], df['b'])
```
